```python
import pandas as pd
import numpy as np


df = pd.read_csv("indian_scholarships.csv")


print("Dataset Info:")
print(df.info())


print("\nNull Values in Each Column:")
print(df.isnull().sum())

print("\nPercentage of Missing Values:")
print((df.isnull().sum() / len(df)) * 100)
```
